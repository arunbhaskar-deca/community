```python
import pandas as pd

df = pd.read_csv('Restacking_columns.csv')

# Create a list of repeated columns

repeated_cols = df.columns[1:7]
repeated_cols


# Count the number of times each column is repeated

col_reps = set()

i = 1
while(i):
    if repeated_cols[0]+'.'+str(i) in df.columns:
        col_reps.add(i)
        i=i+1
    else:
        break

# Create a copy containing the first set of entries that will be in the final table

df_main = df[df.columns[0:7]].copy()
df_main.head()

# Concatenate each group of columns one by one

for i in col_reps:
    df_main = pd.concat([df_main, pd.DataFrame(df[df.columns[7*i+1:7*(i+1)+1]].values, columns=df_main.columns)], axis=0, ignore_index=True)

df_main

```
